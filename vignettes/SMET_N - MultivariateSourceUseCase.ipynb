### Dowloading a set of Databases

In this case, we will download all the databases and then we will append it on a list

```python
arthritis  = pd.read_csv('Arthritis .csv',delimiter=',',index_col='Unnamed: 0').drop(['ID'],axis = 1)
breastcancer = pd.read_csv('Breast_data.csv',delimiter=',',index_col='Unnamed: 0').drop(['Sample code number'],axis = 1).replace('?',np.nan) # in this DB the missing values are represented as '?' 
glass = pd.read_csv('Glass.csv',delimiter=',')
#ionosphere = pd.read_csv('Ionosphere.csv',delimiter=',')
#iris = pd.read_csv('Iris.csv',delimiter=',',index_col='Id')
#wine = pd.read_csv('Wine.csv',delimiter=',').drop(['Wine'],axis = 1)
```

Once we have downloaded all the dataframes we are going to create a list with the dataframes, and a list with the names of the dataframes as well for the legend of the plot

```python
df = list([arthritis,breastcancer,glass])
names = list(['arthritis','breastcancer','glass','ionosphere','iris','wine'])
```

### Discretizing the Data before entropy calculation

The dataframes will be discretized as soon as the sentropylist will be called

### Source Entropies Measures calculation

In this step we will calculate the entropies measurements of all the dataframes on the df list. The dframes will be discretized and the entropies calcualted as if it will be a single case. 

Then it will calculate some variants: 

- type: 'total' returntype: 'Aggegate' (default) : DataFrame with the aggregate total entropies measures of each dataframe
- type: 'total' returntype: 'Individual'         : DataFrame with the dual total entropies measures of each dataframe
- type: 'dual' returntype: 'Aggregate' (default) : List with the total entropies of each dataframe
- type: 'dual' returntype:  'Individual'         : List with the dual entropies of each dataframe
 

```python
edf = sentropieslist(df, names, type = 'total', returntype='individual', base = 2) 
```

### Source Entropies Entropy Triangle Plotting

In this section, we have to differenciate if the sentropylist, return a list of dataframes or a single dataframe. The function etplot, will indentify wether the argument passed to the function edf is a dataframe, or a list of dataframes.

Then it will calculate some variants: 

- 'total' & 'Aggegate' (default) : PLOT DATAFRAME
- 'total' & 'Individual'         : PLOT DATAFRAME
- 'dual'  & 'Aggregate' (default): PLOT LIST OF DATAFRAMES
- 'dual'  &  'Individual'        : PLOT LIST OF DATAFRAMES
 

```python
#edf = sentropieslist(df, names, type = 'total', returntype='individual', base = 2) 
etplot(edf,names,s_mk=170, pltscale=25)
```
