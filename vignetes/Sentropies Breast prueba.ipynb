```python
import random
def random_points(num_points=25, scale=40):
    points = []
    for i in range(num_points):
        x = random.randint(1, scale)
        y = random.randint(0, scale - x)
        z = scale - x - y
        points.append((x,y,z))
    return points
```

```python
points = random_points(30, scale=scale)
```

```python
scale = 40
figure, tax = inverted_ternary.figure(scale=scale)
figure.set_size_inches(10, 10)
tax.set_title("Scatter Plot", fontsize=20)
tax.boundary(linewidth=2.0)
tax.gridlines(multiple=5, color="blue")
# Plot a few different styles with a legend
pointsa = random_points(2, scale=scale)
tax.scatter(pointsa, marker='s', color='red', label="Red Squares")
tax.left_axis_label("Delta", fontsize=20)
tax.right_axis_label("MIP", fontsize=20)
tax.bottom_axis_label("VIP", fontsize=20)
tax.scatter(points, marker='D', color='green', label="Green Diamonds")
tax.legend()
tax.ticks(axis='lbr', linewidth=1, multiple=5)
tax.clear_matplotlib_ticks()

tax.show()
```
