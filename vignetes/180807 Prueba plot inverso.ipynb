```python
from codetriangle.sentropies import *
from codetriangle.coordsentropic import *
from codetriangle.entropytriangle import *
from codetriangle.inverted_ternary import *
from matplotlib import rc

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data', names = ['Sample code number','Clump Thickness','Uniformity of Cell Size','Uniformity of Cell Shape','Marginal Adhesion','Single Epithelial Cell Size','Bare Nuclei','Bland Chromatin','Normal Nucleoli','Mitoses','Class'])
df = df.drop(['Sample code number'],axis = 1)
edf = sentropies_df(df , 'total')
```

```python

scale = 100 ; fonts = 30 ; multiple=5 ; s_mk = 80 ; offset = 0.08
points = entcoords(edf,scale)
figure, tax = inverted_ternary.figure(scale=scale)
figure.set_size_inches(25, 25)
tax.set_title("Source Entropy Multivariate Entropy Triangle (SMET) Breast Cancer", fontsize = fonts)
tax.boundary(linewidth=2.0)
tax.gridlines(multiple=5, color="blue")
# Plot a few different styles with a legend

mk = markers(len(edf.index)) ; colors = get_cmap(len(edf.index));names = list(edf.index)

#colors random
for i in range(len(edf.index)):
    tax.scatter(points[i:i+1], s = 80, marker=mk[i], color= colors(i), label = names[i] ,edgecolor='black', linewidth='3')

    
#tax.scatter(points, marker=',', color='red', label="Red Squares")
#tax.scatter(points, marker='s', color='red', label="Red Squares")
tax.left_axis_label(r"$ VI'_{P_\overline{XY}}$", fontsize=fonts)
tax.right_axis_label(r"$ 2\cdot{I'_{P_\overline{XY}}}  $", fontsize=fonts)
tax.bottom_axis_label(r"$\Delta H'_{P_\overline{XY}}$", fontsize=fonts)
#tax.scatter(points, marker='D', color='green', label="Green Diamonds")
tax.legend()
tax.ticks(axis='lbr', linewidth=1, multiple=5)
tax.clear_matplotlib_ticks()


tax.show()
```
