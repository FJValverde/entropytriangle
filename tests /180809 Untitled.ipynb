```python
from entropytriangle.coordsentropic import *
```

```python
points = entcoords(edf1)
scale = 100
gridl = 5
fonts = 30
offset = 0.02
```

```python
figure, tax = ternary_axes_subplot.figure(scale=scale)
figure.set_size_inches(17, 17)
tax.boundary(linewidth=2.0)
tax.gridlines(multiple = gridl, color="blue")
colors = get_cmap(len(edf1.index)) ; names = list(edf1.index)
if(any(edf1.index == 'AGGREGATE')):
    mk = markers(1)*len(edf1)
    for i in range(len(edf1.index)):
        tax.scatter(points[i:i+1], s = 300, marker = mk[i], color = colors(i), label = names[i] ,edgecolor='black', linewidth='0.5')
        
tax.set_title("Aggregate Channel Multivariate entropies (CMET)", fontsize = fonts + 10)
tax.left_axis_label(r"$ VI'_{P_\overline{XY}}$", fontsize=fonts)
tax.right_axis_label(r"$ 2\cdot{I'_{P_\overline{XY}}}  $", fontsize=fonts)
tax.bottom_axis_label(r"$\Delta H'_{P_\overline{XY}}$", fontsize=fonts , position = (0.5-1.3*offset, 2* offset, 0.5))

tax.ticks(axis='lbr', linewidth=1, multiple = gridl , fontsize = 12 )
tax.clear_matplotlib_ticks()
tax.legend(title = 'Features' ,labelspacing = 1.5 , fontsize = 12)
tax.show()
```
